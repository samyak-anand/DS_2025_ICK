```python
word_freq = {
  'love': 25,
  'conversation': 1,
  'every': 6,
  "we're": 1,
  'plate': 1,
  'sour': 1,
  'jukebox': 1,
  'now': 11,
  'taxi': 1,
  'fast': 1,
  'bag': 1,
  'man': 1,
  'push': 3,
  'baby': 14,
  'going': 1,
  'you': 16,
  "don't": 2,
  'one': 1,
  'mind': 2,
  'backseat': 1,
  'friends': 1,
  'then': 3,
  'know': 2
}
```

```python
# 1.1.1 How many key-value pairs are in the dictionary?
print("Number of key-value pairs:", len(word_freq))
```

```python
# 1.1.2 What keys are present in this dictionary?
print("List of keys:", list(word_freq.keys()))
```

```python
# 1.1.3 Frequency of specific words
print("Frequency of 'friends':", word_freq.get('friends', 0))
print("Frequency of 'taxi':", word_freq.get('taxi', 0))
print("Frequency of 'jukebox':", word_freq.get('jukebox', 0))
```

```python
# 1.1.4 Check if word 'begin' exists in dictionary
print("Is 'begin' in the dictionary?", 'begin' in word_freq)
```

```python
# 1.1.5 Add new words and frequencies
word_freq['begin'] = 1
word_freq['start'] = 2
word_freq['over'] = 1
word_freq['body'] = 17
```

```python
# 1.1.6 Convert dictionary keys to list
word_list = list(word_freq.keys())
print("Word list:", word_list)
```

```python
# 1.1.7 First word in word_list and its frequency
print("First word:", word_list[0], "Frequency:", word_freq[word_list[0]])
```

```python
# 1.1.8 Last word in word_list and its frequency
print("Last word:", word_list[-1], "Frequency:", word_freq[word_list[-1]])
```

```python
# 1.2 Can a dictionary have two key-value pairs with the same key?
# Answer: No. A dictionary cannot have duplicate keys.
# If you assign the same key twice, the latest value will overwrite the previous one.

sample_dict = {'a': 1, 'a': 2}
print(sample_dict)  # Output: {'a': 2}

# 1.3 Can a dictionary have two key-value pairs with the same value but different keys?
# Answer: Yes. Values can be the same for multiple keys.


example_dict = {'a': 10, 'b': 10, 'c': 5}
print(example_dict)  # Output: {'a': 10, 'b': 10, 'c': 5}
```
