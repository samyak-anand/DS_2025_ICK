```python
# 1. Implementing input validation loop

while True:
    user_input = input("Please enter 1 or 2: ")

    # Remove leading spaces
    user_input = user_input.lstrip()

    # Optional: Remove specific unwanted characters (Example - not necessary here)
    user_input = user_input.replace("@", "")  # example of cleaning input

    # Check if input is numeric
    if not user_input.isdigit():
        print("Invalid input! Please enter a number (1 or 2).")
        continue  # go back and ask again

    # Convert to integer
    user_input = int(user_input)

    # Check if input is valid choice
    if user_input == 1 or user_input == 2:
        print(f"Thank you! You selected option {user_input}.")
        break  # exit the loop if input is valid
    else:
        print("Invalid choice! Please enter 1 or 2.")

print("Program ended.")
```
