```python
#Given List
lst = [1, 2, 34, 5, 3, 12, 9, 8, 67, 89, 98, 90, 39, 21, 45, 46, 23, 13]
```

```python
print(" Number of element in the lsit:",len(lst))
```

```python
# 1.2 First element in the list (index 0)
print("First element:", lst[0])
```

```python
# 1.3 Last element in the list (index -1)
print("Last element:", lst[-1])
```

```python
# 1.4 Index of element 90
print("Index of element 90:", lst.index(90))
```

```python
# 1.5 First 8 elements
print("First 8 elements:", lst[:8])
```

```python
# 1.6 Append elements 100 and 110
lst.append(100)
lst.append(110)
print("List after appending 100 and 110:", lst)
```

```python
# 1.7 Sort the list
lst.sort()
print("Sorted list:", lst)
```

```python
names = ["Alice", "Bob", "Charlie", "Dave"]

# Explanation:
# This list 'names' contains 4 string elements.
# We want to search if "Bob" exists in this list and return its index.

search_name = "Bob"

if search_name in names:
    print(f"{search_name} is found at index:", names.index(search_name))
else:
    print(f"{search_name} is not present in the list.")
```

```python
# Example list with positive and negative numbers
numbers = [10, -5, 20, 15, -10, 30]

# Calculate the sum of elements
total = sum(numbers)

# Calculate the length of the list
count = len(numbers)

# Calculate average
average = total / count

print("The average of the list elements is:", average)
```
