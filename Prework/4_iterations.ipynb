```python
# 1.5 Calculate factorial of a number using loop and math library

import math

n = int(input("Enter a positive integer: "))

# Calculate factorial using loop
factorial = 1
for i in range(1, n + 1):
    factorial *= i

print(f"Factorial of {n} using loop is: {factorial}")

# Verify with math library
print(f"Factorial of {n} using math library is: {math.factorial(n)}")
```

```python
x = [12, 43, 4, 1, 6, 343, 10, 34, 12, 93, 783, 330, 896, 1, 55]

# Explanation:
# 'x' is a list containing integer numbers.

# 2.1 Find numbers divisible by 3
print("Numbers divisible by 3:")
for num in x:
    if num % 3 == 0:
        print(num)

# 2.2 Find numbers ending with digit 3
end_with_3 = []
for num in x:
    if str(num)[-1] == "3":  # Convert number to string and check last character
        end_with_3.append(num)

print("Numbers ending with digit 3:", end_with_3)

# Explanation:
# string_value = "IRONHACK"
# string_value[0] gives 'I'
# Similarly, str(num)[-1] gives the last digit of number as string.

# 2.3 Find minimum value in list
print("Minimum value in list:", min(x))

# 2.4 Find maximum value in list
print("Maximum value in list:", max(x))
```

```python
word_freq = {
    'love': 25, 'conversation': 1, 'every': 6, "we're": 1, 'plate': 1, 'sour': 1, 'jukebox': 1, 'now': 11,
    'taxi': 1, 'fast': 1, 'bag': 1, 'man': 1, 'push': 3, 'baby': 14, 'going': 1, 'you': 16, "don't": 2,
    'one': 1, 'mind': 2, 'backseat': 1, 'friends': 1, 'then': 3, 'know': 2
}

# Explanation:
# Dictionary contains words as keys and frequency (integer) as values.

# 3.1 Print keys with frequency < 3
print("Words with frequency less than 3:")
for word, freq in word_freq.items():
    if freq < 3:
        print(word)

# 3.2 Print word with highest frequency
# Approach 1 - Using max function
max_word = max(word_freq, key=word_freq.get)
print(f"Word with highest frequency is '{max_word}' with frequency {word_freq[max_word]}")


```

```python
# Approach 2 - Manual approach
max_freq = 0
word_with_max_freq = ""
for word, freq in word_freq.items():
    if freq > max_freq:
        max_freq = freq
        word_with_max_freq = word

print(f"Highest frequency word is '{word_with_max_freq}' with frequency {max_freq}")
```

```python
import random

# Generate random number between 1 and 100
random_number = random.randint(1, 100)

# Explanation:
# random.randint(a, b) generates a random integer between a and b (inclusive).

# Guessing game
while True:
    guess = int(input("Guess a number between 1 and 100: "))
    if guess > random_number:
        print("Too high, try again!")
    elif guess < random_number:
        print("Too low, try again!")
    else:
        print("Congratulations, you guessed the number!")
        break
```
